```python
import numpy as np
import matplotlib.pyplot as plt

X=[1.3,1.5,2.6,2.6,3.2,3.9,4.2,3.7,3.10,3.8,11]

num_bins=5
count,bin_edges=np.histogram(X,bins=num_bins,density=True)
bin_widths=np.diff(bin_edges)
density=count+bin_widths
print("bin edges : " ,bin_edges)
print("Counts(density values):",count)
print("Density values adjusted for bin width : ",density)

plt.hist(X,bins=num_bins,density=True,edgecolor='red')
plt.xlabel('Data')
plt.ylabel('Density')
plt.title("Histogram with density")
plt.show()
```
